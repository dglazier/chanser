### Import chanser_mva stuff

```python
import ROOT
from ROOT import TMVA, TFile
TMVA.Tools.Instance()
ROOT.TMVA.PyMethodBase.PyInitialize()
ROOT.gSystem.Load("$CHANSER/lib/libchansermva.so")
from ROOT import chanser
from ROOT.chanser import mva
from ROOT.chanser.mva import TrainSignalID 
from ROOT.chanser.mva import Meths
```

### Get the tutorial data file. 
This is the same as is used in the ROOT TMVA classification tutorial

```python
if ROOT.gSystem.AccessPathName( "tmva_class_example.root" ) != 0: 
    ROOT.gSystem.Exec( "wget https://root.cern.ch/files/tmva_class_example.root")
    
input = TFile.Open( "tmva_class_example.root" )

# Get the signal and background trees for training
signalTree      = input.Get( "TreeS" )
backgroundTree  = input.Get( "TreeB" )
```

### Create the training  object. 
You can give any name and this will be the top level directory  for Training.root and weights file.

```python
train = TrainSignalID("TMVAClassificationTut");
```

### Configure the training

Give an output directory destination. Not the name in the cell above will also be added for this instance of the training.

Set any branches in the tree that you which to ignore in the training

Provide the trees we got from the ROOT file above

Set how many events you would like to use from each for testing and training

Prepare the trees for training

```python
train.SetOutDir("");
train.IgnoreBranches("");
train.AddSignalTree(signalTree);
train.AddBackgroundTree(backgroundTree);
train.SetNTrainTest(1500,1500);
train.PrepareTrees();
```

### Configure different classifiers
The arguments here are standard for ROOT TMVA. You will find more details in their documentation

```python
train.BookMethod(TMVA.Types.kBDT, "BDT","!H:!V:NTrees=850:MinNodeSize=2.5%:MaxDepth=3:BoostType=AdaBoost:AdaBoostBeta=0.5:UseBaggedBoost:BaggedSampleFraction=0.5:SeparationType=GiniIndex:nCuts=20");
```

A number of standard configurations are predefined in the MethConf.h file and can directly be used here.

```python
train.BookMethod(Meths.MLP);
```

```python
train.DoTraining();
```
